## 3. Features Understanding

### Perhitungan Banyaknya Items

```python
phone_sale['Brands'].value_counts()
```

### Bar Chart

```python
ax = phone_sale['Brands'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Top 10 Brands Sales')
ax.set_xlabel('Brands')
ax.set_ylabel('Count')
```

### Histogram

```python
ax = phone_sale['Rating'].plot(kind='hist',
                          bins=20,
                          title='Ratings Distribution')
ax.set_xlabel('Ratings')
```

### KDE Plot

```python
ax = phone_sale['Rating'].plot(kind='kde',
                          title='Ratings Distribution')
ax.set_xlabel('Ratings')
```

```python
phone_sale['Mobile'].value_counts()
```

## 4. Feature relationship

### Scatterplot

```python
phone_sale.plot(kind='scatter',
        x='Rating',
        y='Original Price',
        title='Rating vs. Price')
plt.show()
```

```python
ax = sns.scatterplot(x='Rating',
                y='Original Price',
                hue='Brands',
                data=phone_sale)
ax.set_title('Coaster Speed vs. Height')
plt.show()
```

### Correlation & Heatmap

```python
phone_sale_corr = phone_sale[['Rating','Original Price']].dropna().corr()
phone_sale_corr
```

```python
sns.heatmap(phone_sale_corr, annot=True)
```
